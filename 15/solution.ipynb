```python
from collections import *
from heapq import *
from functools import *
from math import *
import numpy as np
import copy

data = None

room_ = []
instructions = []

with open('input.txt', 'r') as file:
    data = file.readlines()
    data = [d.strip() for d in data]
    for d in data:
        if "#" in d:
            room_.append([c for c in d])
        else:
            instructions.append(d)
    instructions = ''.join(instructions)
room = copy.deepcopy(room_)
m,n = len(room),len(room[0])

            
            
print(room)
```

```python
directions = [">","<","^","v"]

def getDirections(x,y):
    return [[x+1,y],[x-1,y],[x,y-1],[x,y+1]]
def printRoom():
    for row in room:
        print(''.join(row))
    print(' ')

printRoom()

def push(x,y,direction): # can i move 1 over
    if not (1 <= x < n-1 and 1 <= y < m-1):
        return False
    cell = room[y][x]
    if cell == "#":
        return False
    if cell == ".":
        return True
    for d,(nx,ny) in zip(directions,getDirections(x,y)):
        if d != direction:
            continue
        if push(nx,ny,direction):
            room[ny][nx] = cell
            room[y][x] = '.'
            return True
        else:
            return False
    print("err")
    return False

ax,ay = 0,0

for y,row in enumerate(room):
    for x,cell in enumerate(row):
        if cell == "@":
            ax,ay =x,y
            break

for c in instructions:
    p = push(ax,ay,c)
    if p:
        for d,(nx,ny) in zip(directions,getDirections(ax,ay)):
            if d != c:
                continue
            ax,ay = nx,ny
    #printRoom()

gpssum = 0
for y,row in enumerate(room):
    for x,cell in enumerate(row):
        if cell == "O":
            gpssum += 100*y + x
print(gpssum)
            
    


    
    
    
```
